```python
name = "ECB_Test_corpus.txt"
with open(name,"r") as file:
    lines = file.readlines()
```

```python
import re
sentences = []
sentence = []
previous_chain = "2_11ecb0"
current_chain = ""
for line in lines:
    if line != '\n':
        tokens = line.split("\t")    
        current_chain = tokens[0]+tokens[1]
        if current_chain!=previous_chain:
            sentence = " ".join(sentence)
#             sentence = re.sub(r'\s([?.!"\',\[\]\(\)](?:\s|$))', r'\1', sentence)
            sentences.append(sentence)
            sentence = []
        sentence.append(tokens[3])
        previous_chain=current_chain
```
